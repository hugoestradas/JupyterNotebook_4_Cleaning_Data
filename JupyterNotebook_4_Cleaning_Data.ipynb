```
missing_values = ["n/a", "na", "--"]
```

```
print(df.head())
```

```
print(df["NUM_BEDROOMS"])
print(df["NUM_BEDROOMS"].isnull())
```

```
print(df["OWN_OCCUPIED"])
print(df["OWN_OCCUPIED"].isnull())
```

```
counter = 0
for row in df["OWN_OCCUPIED"]:
    try:
        int(row)
        df.loc[counter, "OWN_OCCUPIED"]=np.nan
    except ValueError:
        pass
    counter+=1
```
